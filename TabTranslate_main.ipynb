```python
import pandas as pd
import numpy as np
```

```python
#iperparametri#
Nstring = 6 
OrigTun = ['e', 'a' , 'd' , 'g' , 'b' , 'e']
FinTun = ['d', 'a', 'f#', 'd', 'a' , 'd' ]
OrigTon = 'E'
FinTon = 'D'
filename = str('Jhonny_prova_2.txt')


#parametri#
Note=['a', 'a#', 'bb', 'b','cb', 'b#','c' , 'c#' , 'bd' , 'd', 'b#' ,'eb', 'e', 'fb',  'e#', 'f','f#', 'gb',  'g' , 'g#', 'ab']
NoteMap = [0 , 1, 1, 2, 2, 3, 3, 4, 4, 5, 6, 6 , 7 ,7 , 8, 8, 9 , 9, 10, 11, 11  ]
```

```python
def open_file(filename):
    #open the file 
    with open(filename) as f:
        lines = f.readlines()
        #print (lines)
    
    return lines ; 
    
```

```python
def TonDiff(): 
    inVal= NoteMap[Note.index(OrigTon.lower())]
    transVal = NoteMap[Note.index(FinTon.lower())]
    diff = transVal - inVal
    print('moving '+ str(diff) + ' semi tones')
    return diff ; 
```

```python
def find_Tun (lines):

    c = 0 
    tun= []
    for item in lines :
        c=c+1
        if c <= Nstring and (item[0].lower() == 'a' or item[0].lower() == 'b' or item[0].lower() == 'c' or item[0].lower() == 'd' or item[0].lower() == 'e' or item[0].lower() == 'f' or item[0].lower() == 'g') : 
            tun.append(item[0].lower())
    if len(tun) == Nstring : 
        print('the original tuning found in the text is : ') 
        print(tun)
    else : 
        tun = OrigTun 
        print('no tuning used in the text, the specified tuning will be used ')
        print (tun) 
        
    return tun; 
```

```python
def ToneTrans (lines):
    i = 0  
    translines = []
    for line in lines:
        transline=[]
        if line != '\n' and '-' in line:

            for item in line : 
                try : 
                    item = int(item)
                except : 
                    pass 
                if type(item ) == int :
                    transNote =f'{item + diff}'       


                else : 
                    transNote=str(item) 

                transline.append(str(transNote))
            transline = ''.join(transline)
           # print((transline))        
            translines.append(transline)
            #print (i)

            i=i+1
        else : 
            translines.append(line)
        if i >= Nstring: 
            i = 0        
    #print(translines)
    return translines;
```

```python
def Note_Difference(tun , FinTun ): 
    #find the difference between in tuning and new tuning
    i=0
    trans = []
    while i<Nstring :
        for item in tun : 
            notein = item 
            notetrans = FinTun[ i ]
            i=i+1
            inVal= NoteMap[Note.index(notein)]
            transVal = NoteMap[Note.index(notetrans)]
            NoteTrans = inVal - transVal
            trans.append(NoteTrans)

    #print(trans)
    return trans ; 
```

```python
def change_Note_name (lines):
    i=0
    new_lines= []
    for line in lines : 
        if line != '\n' and '-' in line :
            #print (i)

            s= list(line)
            s[0] = FinTun[i]
           # print (s[0])
            line = ''.join(s)
           # print(line)
            if i<Nstring-1 : 
                i=i+1
            else : 
                i=0
        new_lines.append(line)
    #print(new_lines)
    return new_lines ; 
```

```python
def TabTrans (lines):
    i = 0  
    translines = []
    for line in lines:
        transline=[]
        if line != '\n' and '-' in line:
            for item in line : 
                try : 
                    item = int(item)
                except : 
                    pass 
                if type(item ) == int :
                    transNote =f'{item + trans[i]}'       


                else : 
                    transNote=str(item) 

                transline.append(str(transNote))
        else : 
            transline.append(line)
                
        transline = ''.join(transline)
       # print((transline))        
        translines.append(transline)
        #print (i)
        if line != '\n':

            i=i+1
        if i >= Nstring: 
            i = 0        
    #print(translines)
    return translines;
```

```python
def save(filename , file ):
    txt = ['song : '+str(filename) , '\n', '\n' , ' transposed from '+str(OrigTon)+' to '+str(FinTon),'\n', '\n' , 'original tuning : '+ str(OrigTun) , '\n' , 'final tuning : ' + str(FinTun), '\n' , '\n' , '\n']
    txt.extend(file)
    with open(str(filename)+'.txt', 'w') as f:
        for line in txt:
            f.write(line)
            f.write('\n')
```

```python
lines = open_file(filename)



diff = TonDiff()

tun = find_Tun(lines)

trans = Note_Difference (tun , FinTun)



#print (lines) 
toneTrans = ToneTrans(lines)


lines = change_Note_name(toneTrans)


translines=TabTrans(lines)


```
